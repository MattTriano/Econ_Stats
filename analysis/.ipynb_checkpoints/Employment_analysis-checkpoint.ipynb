```python
CSV_PATH = os.path.join('..', 'data_raw', 'manufacturing_employment_CES3000000001.xlsx')
def load_employment_data(CSV_PATH):
    df = pd.read_excel(CSV_PATH, skiprows=12)
    employment = df.melt(id_vars=['Year'])
    employment.dropna(inplace=True)
    employment['value'] = employment['value'].astype('int64')
    employment['Year'] = employment['Year'].astype(str)
    employment['date'] = employment['variable'] + ' 15, ' + employment['Year']
    employment['date'] = pd.to_datetime(employment['date'], format='%b %d, %Y')
    employment = employment.set_index('date').sort_index()
#     employment = employment.reset_index()
    return employment
```

```python
def color_by_president(ax):
    dem_blue = '#0015bc'
    gop_red = '#de0100'
    ax.axvspan(datetime.strptime('Jan 20, 1961', '%b %d, %Y'),
               datetime.strptime('Jan 20, 1969', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 1969', '%b %d, %Y'),
               datetime.strptime('Jan 20, 1977', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 1977', '%b %d, %Y'),
               datetime.strptime('Jan 20, 1981', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 1981', '%b %d, %Y'),
               datetime.strptime('Jan 20, 1993', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 1993', '%b %d, %Y'),
               datetime.strptime('Jan 20, 2001', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 2001', '%b %d, %Y'),
               datetime.strptime('Jan 20, 2009', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 2009', '%b %d, %Y'),
               datetime.strptime('Jan 20, 2017', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 2017', '%b %d, %Y'),
               datetime.strptime('Jan 20, 2021', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    return ax
```

```python
def color_by_senate(ax):
    dem_blue = '#0015bc'
    gop_red = '#de0100'
    ax.axvspan(datetime.strptime('Jan 3, 1955', '%b %d, %Y'),
               datetime.strptime('Jan 3, 1981', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)    
    ax.axvspan(datetime.strptime('Jan 3, 1981', '%b %d, %Y'),
               datetime.strptime('Jan 3, 1987', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)    
    ax.axvspan(datetime.strptime('Jan 3, 1987', '%b %d, %Y'),
               datetime.strptime('Jan 3, 1995', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)    
    ax.axvspan(datetime.strptime('Jan 3, 1995', '%b %d, %Y'),
               datetime.strptime('Jan 3, 2007', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)    
    ax.axvspan(datetime.strptime('Jan 3, 2007', '%b %d, %Y'),
               datetime.strptime('Jan 3, 2015', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)    
    ax.axvspan(datetime.strptime('Jan 3, 2015', '%b %d, %Y'),
               datetime.strptime('Jan 3, 2021', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)    
    return ax
```

```python
def color_by_house(ax):
    dem_blue = '#0015bc'
    gop_red = '#de0100'
    ax.axvspan(datetime.strptime('Jan 3, 1955', '%b %d, %Y'),
               datetime.strptime('Jan 3, 1995', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 3, 1995', '%b %d, %Y'),
               datetime.strptime('Jan 3, 2007', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 3, 2007', '%b %d, %Y'),
               datetime.strptime('Jan 3, 2011', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 3, 2011', '%b %d, %Y'),
               datetime.strptime('Jan 3, 2019', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)    
    ax.axvspan(datetime.strptime('Jan 3, 2019', '%b %d, %Y'),
               datetime.strptime('Jan 3, 2021', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    return ax
```

```python
def format_color_labels(color_labels):
    if len(color_labels) == 3:
        c_label = '\nColored by the Parties of the ' + ', '.join(color_labels[0:2]) + ', and ' + color_labels[2]
    elif len(color_labels) == 2:
        c_label = f"\nColored by the Parties of the {color_labels[0]} and the {color_labels[1]}"
    elif len(color_labels) == 1:
        c_label = f"\nColored by the Party of the {color_labels[0]}"
    else:
        c_label = ''
    return c_label
```

```python
df = mfg_employment
label='Manufacturing Employment (in thousands)'
start_date = '1976-01-01'
end_date = '2021-01-01'
figsize=(12,8)
by_pres=True
by_house=True
by_senate=True

timespan_label=f'{start_date}_to_{end_date}'
fig, ax = plt.subplots(sharex=True, figsize=figsize)
plot_df = df.loc[start_date:end_date, 'value']
ax = plot_df.plot(linewidth=4, color='black',label=label, ax=ax)
ax.tick_params(labelsize=14)
color_labels = []
if by_pres:
    ax = color_by_president(ax=ax)
    color_labels.append('President')
if by_house:
    ax = color_by_house(ax)
    color_labels.append('House')
if by_senate:
    ax = color_by_senate(ax)
    color_labels.append('Senate')
color_label = format_color_labels(color_labels)
    
ax.set_xlim(([start_date, end_date]))
ax.set_ylabel(label, fontsize=18)
ax.set_xlabel('Year', fontsize=18)
ax.set_title(f'{label}{color_label}', fontsize=22)

outpath = os.path.join('..', 'images', f'total_employment_{timespan_label}.png')
plt.savefig(outpath, transparent=False, facecolor='white', edgecolor='black')
```

```python
def make_employment_plot(df, label, start_date='1976-01-01', end_date='2021-01-01',
                         figsize=(12,8), by_pres=True, by_house=True, by_senate=True,
                         zero_anchor=False):
    timespan_label=f'{start_date}_to_{end_date}'
    fig, ax = plt.subplots(sharex=True, figsize=figsize)
    plot_df = df.loc[start_date:end_date, 'value']
    ax = plot_df.plot(linewidth=4, color='black',label=label, ax=ax)
    ax.tick_params(labelsize=14)
    color_labels = []
    if by_pres:
        ax = color_by_president(ax=ax)
        color_labels.append('President')
    if by_house:
        ax = color_by_house(ax)
        color_labels.append('House')
    if by_senate:
        ax = color_by_senate(ax)
        color_labels.append('Senate')
    color_label = format_color_labels(color_labels)
    c_lab = ''.join([el[0].lower() for el in color_labels])
    
    ax.set_xlim([start_date, end_date])
    if zero_anchor:
        ax.set_ylim([0, plot_df.max() * 1.1])
    ax.set_ylabel(label, fontsize=18)
    ax.set_xlabel('Year', fontsize=18)
    ax.set_title(f'{label}{color_label}', fontsize=22)
    
    outpath = os.path.join('..', 'images', f'total_employment_{timespan_label}_{c_lab}.png')
    plt.savefig(outpath, transparent=False, facecolor='white', edgecolor='black')
```

```python
make_employment_plot(df=mfg_employment, label='Manufacturing Employment (in thousands)',
                     start_date='1976-01-01', end_date='2021-01-01', figsize=(12,8),
                     by_pres=True, by_house=True, by_senate=True, zero_anchor=False)
```

```python
def monthly_change(df, start_date, end_date, fs):
    ax = df.loc[start_date: end_date, 'value'].diff().plot(figsize=(14, 8), 
                                                           linewidth=1.5, 
                                                           color='black', 
                                                           label='Employment (in thousands)')
    dem_blue = '#0015bc'
    gop_red = '#de0100'
    ax.tick_params(labelsize=14)
    ax.axvspan(datetime.strptime('Jan 20, 1977', '%b %d, %Y'),
               datetime.strptime('Jan 20, 1981', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 1981', '%b %d, %Y'),
               datetime.strptime('Jan 20, 1989', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 1989', '%b %d, %Y'),
               datetime.strptime('Jan 20, 1993', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 1993', '%b %d, %Y'),
               datetime.strptime('Jan 20, 2001', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 2001', '%b %d, %Y'),
               datetime.strptime('Jan 20, 2009', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 2009', '%b %d, %Y'),
               datetime.strptime('Jan 20, 2017', '%b %d, %Y'), 
               color=dem_blue, lw=1, alpha=0.3)
    ax.axvspan(datetime.strptime('Jan 20, 2017', '%b %d, %Y'),
               datetime.strptime('Jan 20, 2021', '%b %d, %Y'), 
               color=gop_red, lw=1, alpha=0.3)
    ax.set_ylabel('Monthly Change in Total Employment \n(in thousands)', fontsize=fs)
    ax.set_xlabel('Date', fontsize=fs)
    ax.set_xlim([start_date, end_date])
```

```python
monthly_change(employment, start_date='2006-01-01', end_date='2020-12-31', fs=16)
```
